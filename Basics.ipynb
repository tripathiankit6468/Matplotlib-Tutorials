```python
#Import matplotlib 
import matplotlib.pyplot as plt
```

```python
# Creating charts
x = [10,20,30,40,50]
y= [25,50,75,100,125]

plt.plot(x,y)
plt.show()
```

```python
# Customized Chart
plt.figure(figsize=(4,3))
plt.plot(x, y, color = 'green', marker = 'o', linestyle='dashed', linewidth=2, markersize=12)
plt.title("Line Chart")
plt.xlabel("Main X hoon rey")
plt.ylabel("Main Y hoon rey")
plt.show()
```

```python
# Advanced Charts - multiple lines and legends
x = [1,2,3,4,5]
y1 = [10,20,30,40,50]
y2 = [15,25,35,45,55]

plt.plot(x , y1, label='Sales 2024')
plt.plot(x , y2, label='Sales 2025')

plt.title('YoY Sales')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Bar Charts
x = [10,20,30,40,50]
y= [25,50,75,50,25]

plt.bar(x,y)
plt.title('Bar Chart Title')
plt.show()
```

```python
# Data 
import random
import matplotlib.pyplot as plt

data = [random.randint(1, 10) for _ in range(100)]

plt.hist(data, bins=10)
plt.title("Histogram Example")
plt.show()
```

```python
# Pie Chart
categories = ['A','B','C','D','E']
sales= [25,50,75,100,30]
plt.pie(sales,labels=categories,autopct = '%1.1f%%', startangle = 90)
plt.title("Pie Chart Example")
plt.show()
```

```python
# Scatter plot
y1 = [10,20,30,40,50]
y2 = [20,30,40,50,60]

plt.scatter(y1 , y2)
plt.title("Scatter Plot Example")
plt.show()
```

```python
# Subplots
categories = [10,20,30,40,50]
sales = [1000,2000,3000,4000,5000]

y1 = [1,2,3,4,5]
y2 = [20,30,40,50,60]

plt.figure(figsize = (10,6))

# First plot - bar chart
plt.subplot(1,2,1)
plt.bar(categories, sales)
plt.title("Weekly Sales")
plt.xlabel("Week Days")
plt.ylabel("Sales")

# Scatter plot
plt.subplot(1,2,2)
plt.scatter(y1,y1)
plt.title("User Example")
plt.xlabel("User 1")
plt.ylabel("User 2")
plt.show()
```

```python
# Matplotlib with pandas
import pandas as pd
data = {
    'Month' :['Jan','Feb','Mar','Apr'],
    'Sales' : [12000,11000,13000,25000]
    
}
df = pd.DataFrame(data)
df
```

```python
plt.bar(df['Month'], df['Sales'])
plt.title('Matplotlib with pandas')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
# Save charts
plt.bar(df['Month'], df['Sales'])
plt.title('Matplotlib with pandas')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.savefig("Monthly_sales_df.png")
plt.show()
```
