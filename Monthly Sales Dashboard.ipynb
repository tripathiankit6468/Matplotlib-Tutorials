```python
# Import library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load data 
df = pd.read_csv("sales_data.csv")
print(df)
```

```python
# Basic Analysis 
print("Total Sales : ", df["Sales"].sum())
print("Total Profit : ", df["Profit"].sum())
print("Average Sales : " , df["Sales"].mean())
best_month = df.loc[df["Sales"].idxmax()]
print(best_month)
worst_month = df.loc[df["Sales"].idxmin()]
print(worst_month)
```

```python
# Monthly Sales Trend
plt.figure(figsize=(6,5))
plt.plot(df["Month"] , df["Sales"], marker = 'o')
plt.title("Monthly Sales Dashboard")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Monthly figure
plt.figure(figsize = (6,5))
plt.bar(df["Month"], df["Profit"])
plt.title("Monthly Profit")
plt.show()
```

```python
# Customer vs Sales
plt.figure(figsize = (6,5))
plt.scatter(df["Customers"], df["Sales"])
plt.title("Customers Vs. Sales")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate quarterly sales
q1 = df.loc[0:2, "Sales"].sum()    # Jan-Mar
q2 = df.loc[3:5, "Sales"].sum()    # Apr-Jun
q3 = df.loc[6:8, "Sales"].sum()    # Jul-Sep
q4 = df.loc[9:11, "Sales"].sum()   # Oct-Dec

quarters = ["Q1", "Q2", "Q3", "Q4"]
sales = [q1, q2, q3, q4]

plt.figure(figsize=(6,6))

plt.pie(
    sales,
    labels=quarters,
    autopct="%1.1f%%"
)

plt.title("Sales Contribution by Quarter")

plt.show()
```

```python
# Distribution of Monthly Sales
plt.hist(df["Sales"] , bins = 5)
plt.title("Sales Distribution")
plt.show()
```

```python
# Showing sales and profit together
plt.figure(figsize=(10,5))

plt.plot(df["Month"], df["Sales"], marker="o", label="Sales")
plt.plot(df["Month"], df["Profit"], marker="s", label="Profit")

plt.legend()

plt.show()
```

```python
# Highest Sales Month
highest_sales = df.loc[df["Sales"].idxmax()]

# Lowest Sales Month
lowest_sales = df.loc[df["Sales"].idxmin()]

# Highest Profit Month
highest_profit = df.loc[df["Profit"].idxmax()]

# Average Customers per Month
avg_customers = df["Customers"].mean()

print("===== SALES INSIGHTS REPORT =====")

print("\nHighest Sales Month:")
print(highest_sales[["Month", "Sales"]])

print("\nLowest Sales Month:")
print(lowest_sales[["Month", "Sales"]])

print("\nHighest Profit Month:")
print(highest_profit[["Month", "Profit"]])

print("\nAverage Customers per Month:")
print(round(avg_customers, 2))
```
